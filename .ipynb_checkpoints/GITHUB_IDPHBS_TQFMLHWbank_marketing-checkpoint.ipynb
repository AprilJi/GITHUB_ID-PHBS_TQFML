```python
import pandas as pd
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import warnings
```

```python
def versiontuple(v):
    return tuple(map(int, (v.split("."))))


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

## Data processing

```python
#Load the dataset
df = pd.read_csv(U"D:/DataSet1/bank.csv", sep = ";")
df.head()
```

```python
#Choose two futures and split yï¼Œwhere y = 1 if yes, otherwise 0
#For I am not familiar with one-hot encoding, I cannot randomly choose futures, so I only choose age, balance as future examples
ystr = df.loc[:,'y']
X = df.loc[:,['age','balance']]
y = np.where(ystr == 'no', 0, 1)
```

```python
#Split dataset into training/test subsets
from sklearn import datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
#Normalize
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# Full sample
X_combined_std = np.vstack((X_train_std, X_test_std))
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
print(int(0.7*len(y_combined)),len(y_combined))
```

# Using Scikit-learn

```python
method = [" LR ", " SVM ", "Tree", "Forest", "Knn"]
accuracy = np.zeros(5)
```

## 1.Via logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# logistic regression
lr = LogisticRegression(C = 1000, random_state = 0)
lr.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined,
                      classifier=lr, test_idx=range(3164, 4521))
plt.xlabel('age [standardized]')
plt.ylabel('balance [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()

# accuracy of logistic regression
y_pred = lr.predict(X_test_std)
accuracy[0] = round(accuracy_score(y_test, y_pred),4)
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```

## 2.Via support vector machines 

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined,
                      classifier=svm, test_idx=range(3164, 4521))
plt.xlabel('age [standardized]')
plt.ylabel('balance [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()

# accuracy of SVM
y_pred = svm.predict(X_test_std)
accuracy[1] = round(accuracy_score(y_test, y_pred),4)
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```

## 3.Via decision tree learning

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
tree.fit(X_train_std, y_train)


plot_decision_regions(X_combined_std, y_combined, 
                      classifier=tree, test_idx=range(3164, 4521))

plt.xlabel('age [std]')
plt.ylabel('balance [cm]')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()

# accuracy of tree
y_pred = tree.predict(X_test_std)
accuracy[2] = round(accuracy_score(y_test, y_pred),4)
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```

## 4.Via random forests

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini',
                                n_estimators=25, 
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined, 
                      classifier=forest, test_idx=range(3164, 4512))

plt.xlabel('age [std]')
plt.ylabel('balance [cm]')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()

# accuracy of tree
y_pred = forest.predict(X_test_std)
accuracy[3] = round(accuracy_score(y_test, y_pred),4)
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```

## 5.K-nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined, 
                      classifier=knn, test_idx=range(3164, 4512))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()

# accuracy of tree
y_pred = knn.predict(X_test_std)
accuracy[4] = round(accuracy_score(y_test, y_pred),4)
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```
